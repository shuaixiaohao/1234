```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# 1. k值选取不当，导致的错误
train,target = make_blobs(n_samples=150,n_features=2,centers=3)
plt.scatter(train[:,0],train[:,1],c=target)
```

```python
# 一般的k值可以结合业务需求进行制定
kmean1 = KMeans(n_clusters=4)
y1_ = kmean1.fit_predict(train)
plt.scatter(train[:,0],train[:,1],c=y1_)
```

```python
# 2. 数据存在偏差

n = np.array([[1,3],[2,4],[5,6]])
bias = np.array([[0.6,-0.6],[-0.4,0.8]])

np.dot(n,bias)
```

```python
bias_train = np.dot(train,bias)
plt.scatter(bias_train[:,0],bias_train[:,1],c=target)
```

```python
kmean2 = KMeans(n_clusters=3)
y2_ = kmean2.fit_predict(bias_train)
plt.scatter(train[:,0],train[:,1],c=y2_)
```

```python
# 3. 数据方差不一致
train3,target3 = make_blobs(n_samples=150,n_features=2,centers=3,cluster_std=[1,3,5])
plt.scatter(train3[:,0],train3[:,1],c=target3)
```

```python
kmean3 = KMeans(n_clusters=3)
y3_ = kmean3.fit_predict(train3)
plt.scatter(train3[:,0],train3[:,1],c=y3_)
```

```python
# 4. 样本数量不同
train4,target4 = make_blobs(n_samples=1500,n_features=2,centers=3,cluster_std=2)
plt.scatter(train4[:,0],train4[:,1],c=target4)
```

```python
target4 == 0
```

```python
t1 = train4[target4 == 0][:10]
t2 = train4[target4 == 1][:100]
t3 = train4[target4 == 2][:400]
```

```python
n_train = np.vstack((t1,t2,t3))
n_target = np.array([0]*10 + [1]*100 + [2]*400)
```

```python
n_train.shape,n_target.shape
```

```python
plt.scatter(n_train[:,0],n_train[:,1],c=n_target)
```

```python
kmean4 = KMeans(n_clusters=3)
y4_ = kmean4.fit_predict(n_train)
plt.scatter(n_train[:,0],n_train[:,1],c=y4_)
```
