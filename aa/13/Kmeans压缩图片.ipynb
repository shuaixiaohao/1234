```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
girl = plt.imread('goodgirl.jpg')
plt.imshow(girl)
```

```python
# n*3
train = girl.reshape(-1,3)
train.shape
```

```python
girl.shape
```

```python
kmean = KMeans(n_clusters=6)

y_ = kmean.fit_predict(train)
```

```python
# ,
colors = kmean.cluster_centers_
```

```python
# float0~1float32
colors = colors/255
# 0-255uint8
# colors = np.uint8(colors)
```

```python
colors
```

```python
new_image = colors[y_].reshape(girl.shape)
plt.imshow(new_image)
```
