```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
girl = plt.imread('goodgirl.jpg')
plt.imshow(girl)
```

```python
# 对像素点进行聚类分析，所以把样本集改成n*3的格式
train = girl.reshape(-1,3)
train.shape
```

```python
girl.shape
```

```python
kmean = KMeans(n_clusters=6)

y_ = kmean.fit_predict(train)
```

```python
# 获取聚类中心,作为新图片的颜色
colors = kmean.cluster_centers_
```

```python
# 颜色如果是float，必须转换成0~1之间的float32
colors = colors/255
# 颜色如果处于0-255之间的，必须是uint8
# colors = np.uint8(colors)
```

```python
colors
```

```python
new_image = colors[y_].reshape(girl.shape)
plt.imshow(new_image)
```
