```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
# SVC 支持向量机
from sklearn.svm import SVC
```

```python
train = load_iris().data[:,:2]
target = load_iris().target
plt.scatter(train[:,0],train[:,1],c=target)
```

```python
x = np.linspace(train[:,0].min(),train[:,0].max(),200)
y = np.linspace(train[:,1].min(),train[:,1].max(),200)
xx,yy = np.meshgrid(x,y)
X_test = np.c_[xx.ravel(),yy.ravel()]
```

```python
line = SVC(kernel='linear')
rbf = SVC(kernel='rbf') # 默认的内核
poly = SVC(kernel='poly')

line.fit(train,target)
rbf.fit(train,target)
poly.fit(train,target)

results = [line.predict(X_test),rbf.predict(X_test),poly.predict(X_test)]
```

```python
# 绘制图形
from matplotlib.colors import ListedColormap
cmap1 = ListedColormap(['#00aaff','#aaff00','#ff00aa'])
cmap2 = ListedColormap(['#00ff00','#ff0000','#0000ff'])
plt.figure(figsize=(12,4))
titles = ['linear','rbf','poly']
for i,y_ in enumerate(results):
    axes = plt.subplot(1,3,i+1)
    axes.set_title(titles[i])
    axes.scatter(X_test[:,0],X_test[:,1],c=y_,cmap=cmap1)
    axes.scatter(train[:,0],train[:,1],c=target,cmap=cmap2)
```
