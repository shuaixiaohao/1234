```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
train,target = make_blobs(n_samples=160,n_features=2,centers=4)
plt.scatter(train[:,0],train[:,1],c=target)
```

```python
# n_clusters 聚类的个数
kmean = KMeans(n_clusters=4)
# 无监督学习没有标签，只研究数据样本集本身的特征
kmean.fit(train)
# 预测训练集
y_ = kmean.predict(train)
```

```python
# 无监督学习的算法都实现了fit_predict()函数
# 有监督学习的算法都实现了fit()函数和predict()函数
y1_ = kmean.fit_predict(train)
```

```python
# 展示预测结果和真实结果
plt.figure(figsize=(10,4))
axes1 = plt.subplot(1,2,1)
axes1.set_title('Predict')
axes1.scatter(train[:,0],train[:,1],c=y_)

axes2 = plt.subplot(1,2,2)
axes2.set_title('True')
axes2.scatter(train[:,0],train[:,1],c=target)

# 绘制聚类中心
centers = kmean.cluster_centers_
axes1.scatter(centers[:,0],centers[:,1],c='red',s=80)
```
