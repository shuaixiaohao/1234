```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Ridge,Lasso,LinearRegression
```

```python
n_samples,n_features = 50,200

X = np.random.random(size=(n_samples,n_features))

coefs = np.random.random(size=200)
# 假设200个数据特征只有10个是对结果有影响
index = np.arange(0,200,1)
# 修改输入数据的内存结构
np.random.shuffle(index)
```

```python
coefs[index[10:]] = 0
```

```python
# 系数乘以样本特征，就得到了样本标签
y = np.dot(X,coefs)
y.shape
```

```python
y.max(),y.min()
```

```python
# 对样本标签添加噪声
noise = np.random.random(size=50)*2-1
noise.max(),noise.min()
```

```python
# 把噪音数据添加到样本标签y中
y += noise
```

```python
plt.plot(coefs)
plt.xlabel('features')
plt.ylabel('coefs')
```

```python
linear = LinearRegression()
ridge = Ridge(alpha=100000)
lasso = Lasso(alpha=0.05)

linear.fit(X,y)
ridge.fit(X,y)
lasso.fit(X,y)

# 保留三种算法的系数
coefs1 = linear.coef_
coefs2 = ridge.coef_
coefs3 = lasso.coef_

plt.figure(figsize=(12,8))
axes1 = plt.subplot(2,2,1)
axes1.plot(coefs)
axes1.set_title('True',color='blue')

axes2 = plt.subplot(2,2,2)
axes2.plot(coefs1,color='green')
axes2.set_title('LinearRegression')

axes3 = plt.subplot(2,2,3)
axes3.plot(coefs2,color='orange')
axes3.set_title('Ridge')

axes4 = plt.subplot(2,2,4)
axes4.plot(coefs3,color='cyan')
axes4.set_title('Lasso')
```
