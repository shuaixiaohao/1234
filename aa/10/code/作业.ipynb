```python
def trans_objdata(columnname):
    unique_arr = train[columnname].unique()

    def map_function(item):
        return np.argwhere(unique_arr == item)[0,0]

    train[columnname] = train[columnname].map(map_function)
```

```python
trans_objdata('occupation')
trans_objdata('sex')
```

```python
# 初步完成数据值的映射
train
```

```python
# 算法选择
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# 拆分训练集和测试集
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(train,target,test_size=0.2,random_state=1)
```

```python
# 训练
knn.fit(X_train,y_train)
```

```python
y_ = knn.predict(X_test)
```

```python
(y_ == y_test).sum()/y_.size
```
