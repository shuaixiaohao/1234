```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
%matplotlib inline

# 逻辑斯蒂函数
from sklearn.linear_model import LogisticRegression
# 随机生成分类样本集的函数
from sklearn.datasets import make_blobs
```

```python
# 返回数据样本和样本标签
train,target = make_blobs(n_samples=150,n_features=2,centers=[[1,3],[5,1],[5,5]])
```

```python
train.shape,target.shape
```

```python
#
plt.scatter(train[:,0],train[:,1],c=target)
plt.xlabel('featrue1')
plt.ylabel('featrue2')
```

```python
# 回归模型，处理分类问题
logistic = LogisticRegression()

logistic.fit(train,target)
```

```python
# 获取所有的预测点
xmin,xmax = train[:,0].min()-0.5,train[:,0].max()+0.5
ymin,ymax = train[:,1].min()-0.5,train[:,1].max()+0.5

x = np.linspace(xmin,xmax,300)
y = np.linspace(ymin,ymax,300)

xx,yy = np.meshgrid(x,y)

X_test = np.c_[xx.ravel(),yy.ravel()]

y_ = logistic.predict(X_test)

plt.scatter(X_test[:,0],X_test[:,1],c=y_)
```
